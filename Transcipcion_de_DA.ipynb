**48. Población, muestra y variables**

**SABERES DINÁMICOS**

La estadística es una rama de la matemática que recolecta, describe, organiza y analiza datos con el fin de interpretar una situación o predecir algo en una cierta población. Los datos se obtienen mediante instrumentos de recolección como encuestas, entrevistas, observaciones, cuestionarios, experimentos, etc.

La población es el conjunto de individuos o elementos que son objeto de estudio (personas, animales, plantas, objetos, etc.). Si la población es muy numerosa, se selecciona una parte de la misma que se denomina muestra, la cual debe ser representativa (con características similares a la población). Esta muestra se selecciona para investigar uno o varios temas de esa población que se denominan variables.

**Tipos de variables**

**Cualitativas:** expresan características o cualidades de la población y no se pueden medir con números.

- **Por ejemplo:** deportes que practican los alumnos de un curso.

**Cuantitativas:** se expresan con datos numéricos.

- **Discretas:** puede tomar una cantidad finita de valores (número contable de valores).
  - **Por ejemplo:** número de calzado de los jugadores de fútbol de un club.

- **Continuas:** puede tomar valores infinitos y por eso se agrupan en intervalos de clases. Estos valores tienen una mínima amplitud.
  - **Por ejemplo:** el peso de los alumnos de un curso.

---

**Variables:** en este caso hay dos:

- **Cantidad de hijos** es una variable cuantitativa discreta, ya que toma una cantidad finita de valores.
- **Estado civil** es una variable cualitativa.

**SABERES DINÁMICOS**

De cada valor de una variable estadística se puede calcular diferentes frecuencias. Las más conocidas son:
- **Frecuencia absoluta (f):** es la cantidad de veces que se repite cada valor de la variable.
- **Frecuencia relativa (fr):** es el resultado de dividir la frecuencia absoluta por el número de elementos del total (N).
- **Frecuencia relativa porcentual (fr%):** es la frecuencia relativa multiplicada por 100.
- **Frecuencia absoluta acumulada (Fa):** es la suma de las frecuencias absolutas de todos los valores menores o iguales a un cierto valor. Por este motivo, solo se puede calcular en variables cuantitativas o cualitativas que admitan una relación de orden. Por ejemplo, la variable calificación de una obra es cuantitativa, ya que la misma admite un orden.

Estos datos se pueden organizar en tablas para facilitar el análisis de la información.

En una panadería se registraron los datos de tortas que compró cada cliente durante el día domingo y los datos fueron volcados en la siguiente tabla:

| Cantidad de tortas compradas por cliente | f (Cantidad de clientes) | fr (frecuencia relativa) | fr% (frecuencia relativa porcentual) | fa (frecuencia absoluta acumulada) |
|-----------------------------------------|--------------------------|--------------------------|-------------------------------------|------------------------------------|
| 0                                       | 18                       | 0.6                      | 60%                                 | 18                                 |
| 1                                       | 4                        | 0.13                     | 13%                                 | 22                                 |
| 2                                       | 3                        | 0.1                      | 10%                                 | 25                                 |
| 3                                       | 3                        | 0.1                      | 10%                                 | 28                                 |
| 4                                       | 2                        | 0.07                     | 7%                                  | 30                                 |
| **Totales**                             | **30**                   | **1**                    | **100%**                            |                                    |

En estadística existen diversas medidas que permiten estudiar características de una variable. Los números índice son un ejemplo de ellas; se utilizan para medir y estudiar los cambios de una variable en el tiempo. Se calcula dividiendo el periodo que deseamos analizar por el periodo base y multiplicando por 100, el periodo inicial, que se denomina periodo base.

Número índice = (Periodo deseado / Periodo base) x 100

Teniendo en cuenta que el número índice no puede ser cero.

En la panadería también se registraron las ventas anuales de la torta brownie y del lemon pie:

| Producto | Año 2018 | Año 2019 | Número índice | Variación % |
|----------|----------|----------|---------------|-------------|
| Torta brownie | 2746 | 2602 | 2746 / 2602 * 100 = 105.5 | +5.53% |
| Lemon pie | 2500 | 1089 | 1089 / 2500 * 100 = 43.56 | -56.44% |

El porcentaje de variación (variación %) es el indicador que puede ser negativo o positivo y se calcula así:

Porcentaje de variación = (Periodo deseado - Periodo base) / Periodo base * 100

En el caso de las ventas de la torta brownie:

Porcentaje de variación = (2746 - 2602) / 2602 * 100 = +5.53%, la variación es positiva.

En el caso del lemon pie:

Porcentaje de variación = (1089 - 2500) / 2500 * 100 = -56.44%, la variación es negativa.

**49. Frecuencias y gráficos**

La información que surge de un estudio estadístico puede ser representada en tablas, o bien utilizando un gráfico, el cual permite visualizar con mayor facilidad la información. En el caso de trabajar con variables cualitativas o con variables cuantitativas discretas se pueden utilizar representaciones gráficas como el gráfico de barras, el circular y el pictograma.

En una competencia de natación se registró la cantidad de estudiantes que participaron diferenciados por año escolar.

| Año | 1° | 2° | 3° | 4° | 5° | Total |
|-----|----|----|----|----|----|-------|
| f   | 6  | 3  | 6  | 9  | 6  | 30    |
| fr  | 0.2| 0.1| 0.2| 0.3| 0.2| 1     |
| fr% | 20%| 10%| 20%| 30%| 20%| 100%  |

---

**Gráfico de barras:** se muestran las frecuencias absolutas de cada valor de la variable. Todos los rectángulos del gráfico tienen el mismo ancho y las alturas coinciden con las frecuencias absolutas de cada dato.

**Gráfico circular:** se muestran las frecuencias relativas porcentuales de cada dato. Este gráfico está dividido en sectores. Para calcular el ángulo central de un sector, por ejemplo, para el dato 1°, se realiza el siguiente cálculo:


100% → 360º

20% → x

$x = \frac{20\% \cdot 360^\circ}{100\%}$

x = 72º


**Pictograma:** las frecuencias absolutas se representan con dibujos relacionados con el tema. Se toma como referencia un dibujo que representa una determinada cantidad.

**50. Medidas de posición**

**SABERES DINÁMICOS**

Las medidas de posición nos permiten conocer diversas características de los datos recolectados de la variable estadística a analizar. Entre estas medidas se encuentran: el promedio o media, la mediana y la moda.

Para calcular el **promedio o media** (se escribe \(\bar{x}\)) se realiza la suma de todos los valores de la variable y se divide por la cantidad de elementos de la muestra.

**En un parque de diversiones se registraron las edades de los ingresantes al Laberinto del terror de la función de las 16 horas:**

Ordenados de menor a mayor:

\[18, 18, 20, 22, 30, 34, 36, 38\]

\[
\bar{x} = \frac{18 + 18 + 20 + 22 + 30 + 34 + 36 + 38}{8} = \frac{216}{8} = 27
\]

**La moda** (se escribe \(m_o\)) es el valor de la variable que tiene mayor frecuencia. Puede haber más de una moda en una variable. En el ejemplo anterior, \(m_o = 18\).

**La mediana** (se escribe \(m_e\)) es el valor de la variable que ocupa el lugar central. Para obtenerla, primero hay que ordenar todos los valores de menor a mayor.

\[18, 18, 20, 22, 30, 34, 36, 38\]

\[
m_e = \frac{22 + 30}{2} = \frac{52}{2} = 26
\]

Si la cantidad de los elementos de la muestra es par, la mediana es el promedio de los dos valores centrales.

**También se registró las edades de los ingresantes de la función de las 22 horas:**

\[18, 26, 28, 30, 36, 40, 42, 48\]

\[
m_e = \frac{30 + 36}{2} = \frac{66}{2} = 33
\]
